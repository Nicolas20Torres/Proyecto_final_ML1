## Descripción del Conjunto de Datos de Obesidad
Este conjunto de datos está diseñado para estudiar los factores de riesgo y las características relacionadas con la obesidad en una población. Los datos incluyen variables demográficas, de estilo de vida y comportamientos alimentarios.  A continuación se presenta una síntesis de las principales características del conjunto de datos:

1. **Información Demográfica:** Incluye variables como el género y la edad, lo cual permite analizar tendencias en distintos grupos de población.

2. **Datos Antropométricos:** Variables como `Height` (altura) y `Weight` (peso) son fundamentales para calcular el Índice de Masa Corporal (`IMC`) y clasificar los niveles de obesidad.

3. **Antecedentes de Salud:** La variable `family_history_with_overweight` indica si los individuos tienen antecedentes familiares de sobrepeso, un factor de predisposición genética relevante.

4. **Hábitos Alimenticios:** Variables como `FAVC` (consumo frecuente de alimentos calóricos), `CAEC` (frecuencia de comer entre comidas) y `NCP` (número de comidas diarias) describen patrones de alimentación.

5. **Estilo de Vida:** Incluye datos sobre el consumo de alcohol (`CALC`), uso de tecnología (`TUE`), y nivel de actividad física semanal (`FAF`), los cuales influyen en la salud y el peso corporal.

6. **Medios de Transporte:** La variable `MTRANS` describe el tipo de transporte utilizado, lo que puede estar relacionado con el nivel de actividad física diario.

7. **Clasificación de Obesidad:** La columna `NObeyesdad` clasifica a los individuos en distintos niveles de obesidad o peso saludable, proporcionando una etiqueta que puede utilizarse en modelos predictivos o para segmentar la población en el análisis.

> En resumen, estos datos permiten investigar cómo los factores de comportamiento, los antecedentes familiares y el estilo de vida contribuyen al riesgo de obesidad en diferentes personas.

## 2. Tipos de Variables en el Conjunto de Datos

Las variables en este conjunto de datos se pueden clasificar en las siguientes categorías:

- **Variables Categóricas Binarias**: Solo tienen dos valores posibles, representando un "sí" o "no" en ciertas características. Ejemplos: `family_history_with_overweight`, `SMOKE`, y `SCC`.

- **Variables Categóricas Nominales**: No tienen un orden inherente. Ejemplos: `Gender`, `MTRANS`, y `NObeyesdad`, donde cada categoría es única y no tiene jerarquía entre ellas.

- **Variables Categóricas Ordinales**: Tienen un orden implícito. Aunque no es evidente en todos los casos, variables como `CAEC` y `CALC` podrían considerarse ordinales en función de la frecuencia, aunque el orden específico debería validarse.

- **Variables Numéricas Continuas**: Pueden tomar una amplia gama de valores decimales y representan mediciones. Ejemplos: `Height`, `Weight`, `CH2O`, y `Age`.

- **Variables Numéricas Discretas**: Solo pueden tomar valores enteros, como `NCP` (número de comidas) y `FAF` (actividad física semanal), aunque en algunos casos pueden incluir valores decimales.

Este análisis ayuda a determinar los métodos de procesamiento y análisis adecuados para cada tipo de variable, facilitando una correcta interpretación y modelado de los datos.

# Análisis de Información Mutua de Variables Categóricas

## ¿Qué es el Análisis de Información Mutua?

El análisis de información mutua es una técnica utilizada para medir la dependencia entre dos variables. En el contexto de machine learning, se utiliza para determinar cuánto conocimiento se puede obtener sobre la variable objetivo al conocer el valor de una variable predictora. La información mutua está en el rango de 0 a 1:

- **0**: Indica que no hay información compartida entre la variable predictora y la variable objetivo, por lo que la variable predictora es irrelevante para el modelo.
- **1**: Indica una dependencia total entre la variable predictora y la variable objetivo, lo que significa que conocer la variable predictora permite predecir la variable objetivo sin ninguna incertidumbre.

Este análisis es especialmente útil para seleccionar características relevantes en modelos de clasificación, ayudando a decidir qué variables son más informativas y deberían ser incluidas en el modelo.

```python
# Instancia de models para llamar analsis de informacion mutua
M = Models(df.df)
col = [
    'CAEC',
    'family_history_with_overweight',
    'CALC',
    'MTRANS',
    'FAVC',
    'SCC',
    'SMOKE']
M.informacion_mutua(col,'NObeyesdad')
```

## Interpretación de los Resultados

### Valores de Información Mutua

- **Información mutua alta**: Un valor alto de información mutua indica que la variable tiene una alta dependencia con la variable objetivo. Esto sugiere que la variable es relevante para la clasificación o predicción y debería considerarse seriamente para incluirla en el modelo.
  
- **Información mutua baja**: Un valor bajo de información mutua indica que la variable aporta poca información sobre la variable objetivo y, por lo tanto, podría ser menos relevante para el modelo.
